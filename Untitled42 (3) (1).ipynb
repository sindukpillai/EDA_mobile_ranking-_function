```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
#read the dataset
dataset=pd.read_csv("C:\\Users\\CSSTF107\\Desktop\\MobileTrain.csv")
dataset.head()
dataset.describe(include='all')
dataset.info()
#Chk for null values present in dataset
dataset.isnull().sum()
```

```python
#countplot for all columns in which number of items are limited
for i in dataset:
    if (dataset[i].nunique()<30):
        sns.countplot(x=dataset[i])
        plt.show()
```

```python
#distplot for numrical features
def draw_distplot(col,x):
    
    plt.subplot(x)
    sns.distplot(dataset[col])
    return
col=['battery_power','clock_speed','int_memory','m_dep','mobile_wt','px_height','px_width','ram','talk_time']
x=330
plt.figure(figsize=(30,10))
for i in range(0,len(col)): 
    x+=1
    draw_distplot(col[i],x)
plt.show()

```

```python
#pie for binary columns in percentage
def create_pie_chart_columns(newcol,col):

    dataset[newcol]=' '
    for i in range(len(dataset)):
        if dataset[col][i]==0:
            dataset[newcol][i]='No'
        else:
            dataset[newcol][i]='Yes'
    
    
    fig1=px.pie(data_frame=dataset,names=newcol,title='percentage of devices having '+newcol,hole=.2)
    fig1.show()
    return
newcol=['isBluetooth','isDualsim','isFourg','isThreeg','isTouchscreen','isWifi']
col=['blue','dual_sim','four_g','three_g','touch_screen','wifi']
for i in range(0,len(newcol)):
    create_pie_chart_columns(newcol[i],col[i])


```

```python
def processor_core():
    dataset['Cores']=' '
    for i in range(len(dataset)):
        if dataset['n_cores'][i]==1:
            dataset['Cores'][i]='Single Core'
        elif dataset['n_cores'][i]==2:
            dataset['Cores'][i]='Dual Core'
        elif dataset['n_cores'][i]==3:
            dataset['Cores'][i]='Trible Core'
        elif dataset['n_cores'][i]==4:
            dataset['Cores'][i]='Quad Core'
        elif dataset['n_cores'][i]==5:
            dataset['Cores'][i]='Penta Core'
        elif dataset['n_cores'][i]==6:
            dataset['Cores'][i]='Hexa Core'
        elif dataset['n_cores'][i]==7:
            dataset['Cores'][i]='Hepta Core'
        elif dataset['n_cores'][i]==8:
            dataset['Cores'][i]='Octa Core'
    return
processor_core()
px.pie(data_frame=dataset,names='Cores',title='percentage of Processor cores',hole=.2)
```

```python
#histogram for binary features  against dependant feature
def draw_hist(newcol,col):
    fig1=px.histogram(data_frame=dataset,x=newcol,color=col,title='comp of dvices having  or not  '+newcol)
    fig1.show()
    return
newcol=['isBluetooth','isDualsim','isFourg','isThreeg','isTouchscreen','isWifi','Cores']
col='price_range'
for i in range(0,len(newcol)):
   
    draw_hist(newcol[i],col)


```

```python
dataset.drop(['isBluetooth','isDualsim','isFourg','isThreeg','isTouchscreen','isWifi','Cores'],axis=1,inplace=True)
x=dataset.iloc[:,:19].values

y=dataset.iloc[:,20].values
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
log=LogisticRegression()
log.fit(x_train,y_train)
print("training scote is {}".format(log.score(x_train,y_train)*100))
y_predlog=log.predict(x_test)
print(y_predlog)
am_log=accuracy_score(y_test,y_predlog)*100
print("accuracy of log regession is {}".format(am_log))
print("confusion matrix is {}".format(confusion_matrix(y_test,y_predlog)))
print("{}".format(classification_report(y_test,y_predlog)))
```
